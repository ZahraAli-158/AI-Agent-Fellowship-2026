# Introduction

Prompt engineering is a technique used to communicate effectively with AI systems by designing clear and purposeful prompts. In this report, different prompting methods are explored to examine their impact on response accuracy, reasoning ability, and overall performance across various tasks.

# Overview

This report presents five practical experiments that explore fundamental prompt engineering techniques, including Role Prompting, Chain-of-Thought Prompting, Few-Shot Prompting, Structured Output (JSON), and Prompt Optimization. Each experiment includes the prompt used, the AI-generated response, an evaluation of the results, and recommendations for further improvement.

Adding another few-shot example with a more confusing email can help the model better understand the difference between Spam and Not Spam, making its predictions more reliable.
